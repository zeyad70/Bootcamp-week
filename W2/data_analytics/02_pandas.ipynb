# Pandas Fundamentals

**Pandas** provides fast, flexible, and expressive data structures designed to make working with “relational” or “labeled” data both easy and intuitive. It is built on top of **NumPy**, which is built on top of **C**. So, it is really fast.

We start with the [**Getting started tutorials**](https://pandas.pydata.org/docs/getting_started/intro_tutorials/index.html). Recommended delivery in two sessions: (~40 min each). To get an overview with real-data examples and visuals.

Know SQL? take a look at: [SQL to Pandas](https://pandas.pydata.org/docs/getting_started/comparison/comparison_with_sql.html).
