## Getting Started

````python
from sklearn.tree import DecisionTreeClassifier

# Model
# this model is used for learning mapping between features (of any type)
# and a categorical target (e.g. "0" or "1")
estimator = DecisionTreeClassifier(random_state=0)

# Data
X = [[ 1,  2,  3],  # 2 samples, 3 features
     [11, 12, 13]]

y = [0, 1]  # classes of each sample

# Train (to find the best mapping: X -> y)
estimator.fit(X, y)
````

````python
# After the model has trained, we want to make sure that if we pass in the same training data
# we actually get the labels (targets) it has been trained on
estimator.predict(X)
````

````python
# We can also use it to predict on new samples
# Let's make up some data:
X_new = [
    [4, 5, 6],   # 0?
    [14, 15, 16] # 1?
]
estimator.predict(X_new)
````

````python
from sklearn.datasets import load_iris

from sklearn.preprocessing import StandardScaler

# from sklearn.linear_model import LinearRegression # Regression Task (Continuous Output)
from sklearn.linear_model import LogisticRegression # because we have 2 classes (Classification Task) (Binary Output)

from sklearn.pipeline import make_pipeline
````

````python
# load the iris dataset and split it into train and test sets
X, y = load_iris(return_X_y=True) # Like: pd.read_csv('data.csv')
````

````python
X.shape
````

````python
y.shape
````

````python
from sklearn.model_selection import train_test_split

# Then, we split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
````

````python
X_train.shape
````

````python
y_train.shape
````

````python
X_test.shape
````

````python
X.shape[0] == X_train.shape[0] + X_test.shape[0]
````

````python
# create a pipeline object
pipe = make_pipeline(
    StandardScaler(),
    LogisticRegression()
)
# fit the whole pipeline
pipe.fit(X_train, y_train)
````

````python
y_pred = pipe.predict(X_test)
y_pred
````

````python
y_pred
````

````python
y_test
````

````python
from sklearn.metrics import (
    accuracy_score, # Classification Target (Categorical Output)
    mean_absolute_error, mean_squared_error # Continuous Target (Regression)
)
accuracy_score(y_pred, y_test)
````
