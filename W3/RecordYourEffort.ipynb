# Record Your Efforts with Git Commits

Each exercise we want a notebook with the name: `Exercise Mx.yy.ipynb`.

1. First commit will contain the exercise (no solution yet)
2. Next commits will contain incremental steps towards the solution (everytime you solve a part of the exercise, commit your code)

The instructor asks you to solve `Exercise M1.01` for the module named `"Tabular data exploration"`. You create a file named `Exercise M1.01.ipynb` and copy the exercise (no solution yet) into that file. And commit that into `git`, with the message:


```sh
git commit -m "add Exercise M1.01: Tabular data exploration"
```

Then, you solve the first part which has the questions: "What are the features? What is the target?" and commit your solution with the message:

```sh
git commit -m "solve M1.01: What are the features? What is the target?"
```

You move on to the next question to solve the question: "How many features are numerical? How many features are categorical?" and commit your solution with the message:

```sh
git commit -m "solve M1.01: How many features are numerical? How many features are categorical?"
```

..etc.

After you have finished the execise, you may push to GitHub, or keep going until the end of the module, then do one single push for all. Either way, the push command is simply:

```sh
git push
```

Note: the text of commit message doesn't have to be exact. But, try to follow the example as much as possible.
